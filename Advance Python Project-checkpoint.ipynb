 # "Handling Missing Data using Imputation Techniques"



```python
df.info()
```

```python
df.isnull().sum()
```

```python
# here the column which  contain  null values are of object datatypes 
```

```python
for i in df.select_dtypes(include='object').columns:
    df[i].fillna(df[i].mode()[0],inplace=True)
```

```python
df.isnull().sum
```

```python
#we sucessfully able to fill missing values 
```

 # Improving Data Uniformity through Date and Number Format Standardization


```python
df.info()
```

```python
# ORDERDATE column should be in date format but it is in object
```

```python
df['ORDERDATE']= pd.to_datetime(df['ORDERDATE'])
```

```python
# drive year from ORDERDATE column
df['YEAR']= df['ORDERDATE'].dt.year
df['YEAR']= df['YEAR'].round().astype(int)
```


 # Statistical and Machine Learning Approaches for Managing Data Gaps and Anomalies



## univariate analysis

```python
sns.countplot(x = df['DEALSIZE'],palette=['green', 'red', 'purple'])
plt.legend(title="DEALSIZE", labels=df['DEALSIZE'].unique())
plt.show()
```

```python
 # Moderate Deal Size Leading to High Revenue
```

```python
plt.pie(df['STATUS'].value_counts(),startangle=90,counterclock=False,wedgeprops={'edgecolor':'black',},colors = ['#ff5733', '#33ff57', '#3357ff', '#f0e68c', '#8a2be2']
)
plt.title("STATUS MODE")
plt.legend(df['STATUS'].unique())
plt.show()
```

```python
# shipped mode have best performance 
```

## Bivariate analyis

```python
plt.bar(df['DEALSIZE'], df['SALES'],color='green')  
plt.xlabel('Deal Size')
plt.ylabel('Sales')
plt.title('Sales by Deal Size')
plt.xticks(rotation=45) 
plt.show()
```

```python
# large deal sixe have high sales
```

```python
plt.bar(df['STATUS'], df['SALES'],color='lightgreen')  
plt.xlabel('Status Mode')
plt.ylabel('Sales')
plt.title('Sales by Status Mode')
plt.xticks(rotation=45) 
plt.show()
```

```python
# in 2005 year company generate high revenue
```

```python
yearly_sales = df.groupby('YEAR')['SALES'].sum().reset_index()
sns.lineplot(x=yearly_sales['YEAR'], y=yearly_sales['SALES'], marker='o', color='blue')
plt.xlabel('Year')
plt.xticks(rotation=90)
plt.ylabel('Total Sales')
plt.title('Sales Trend Over the Years')
plt.show()
```

```python
status_quantity = df.groupby('STATUS')['QUANTITYORDERED'].sum()
plt.bar(status_quantity.index, status_quantity.values)

plt.xlabel('Order Status')
plt.ylabel('Total Quantity Ordered')
plt.title('Quantity Ordered by Order Status')
plt.xticks(rotation=45) 
plt.show()
```

```python
## multivariate analysis
```

```python
status_dealsize_sales = df.groupby(['STATUS', 'DEALSIZE'])['SALES'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='STATUS', y='SALES', hue='DEALSIZE', data=status_dealsize_sales, palette=['green','red', 'pink'])
plt.xlabel('Order Status')
plt.ylabel('Total Sales')
plt.title('Sales by Status and Deal Size')
plt.xticks(rotation=45)
plt.legend(title='Deal Size')
plt.show()
```

```python
# In shipped mode medium dealsize performance is good
```

```python
status_dealsize_QuantityOrder = df.groupby(['STATUS', 'DEALSIZE'])['QUANTITYORDERED'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='STATUS', y='QUANTITYORDERED', hue='DEALSIZE', data=status_dealsize_QuantityOrder,palette=['pink','yellow','skyblue'])
plt.xlabel('Order Status')
plt.ylabel('Total QuantityOrder')
plt.title('QuantityOrder by Status and Deal Size')
plt.xticks(rotation=45)
plt.legend(title='Deal Size')
plt.show()
```

 #  Sales Insights: Trends, Seasonal Patterns, and Best-Performing Products


```python
yearly_sales = df.groupby('YEAR')['SALES'].sum().reset_index()
sns.lineplot(x=yearly_sales['YEAR'], y=yearly_sales['SALES'], marker='o', color='blue')
plt.xlabel('Year')
plt.xticks(rotation=90)
plt.ylabel('Total Sales')
plt.title('Sales Trend Over the Years')
plt.show()
```

```python
# sales fist increase after that it deacrease overe year
```

```python
yearly_sales = df.groupby('YEAR')['QUANTITYORDERED'].sum().reset_index()
sns.lineplot(x=yearly_sales['YEAR'], y=yearly_sales['QUANTITYORDERED'], marker='o', color='red')
plt.xlabel('Year')
plt.xticks(rotation=90)
plt.ylabel('Total QuantityOrder')
plt.title('QuantityOrder Over the Years')
plt.show()
```

```python
top_states = df.groupby('STATE')['SALES'].sum().nlargest(10).reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='STATE', y='SALES', data=top_states, palette='viridis')

plt.xlabel('State')
plt.ylabel('Total Sales')
plt.title('Top 10 States with Highest Sales')
plt.xticks(rotation=45)  
plt.show()
```
